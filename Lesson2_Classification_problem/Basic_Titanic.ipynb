## Data Cleaning

Missing Values
1. Cabin : missing 77% --> not useful features (Drop this rows)
2. Age : missing 20% --> we should try to fill following some strategy in order to apply the filling closer to what would be
3. Embarked : missing 0.2% --> can fill missing values or drop missing rows ( No effect)

Categorical Values
1. Sex --> 0 (Males), 1( Females)
2. Name --> This property doesn't give useful ( Drop column)
3. Ticket --> This property doesn't give useful ( Drop column)
4. Cabin --> missing 77% (Drop column)
5. Embarked --> 3 unique values (one-hot encoder)

```python
train_df['Embarked'].value_counts()
```

###Clean data

```python
def clean_data(data):
    
    # 1. drop not useful column
    data.drop(columns = ['Name','Ticket','Cabin'], inplace = True)

    # 2. fillna on 'Age' with median_age
    median_age = train_df['Age'].median()
    data['Age'] = data['Age'].fillna(median_age)

    # 3. fillna on 'Fare' with median_fare
    median_fare = train_df['Fare'].median()
    data['Fare'] = data['Fare'].fillna(median_age)

    # 4. drop row for missing 'Embarked'
    # axis 0 = row, axis 1 = columns
    data.dropna(axis=0, subset=['Embarked'], inplace=True)

    # 5. One-hot Encoder
    one_hot_columns = pd.get_dummies(data[['Sex','Embarked']])
    data.drop(columns = ['Sex','Embarked'], inplace=True)
    data = pd.concat([data, one_hot_columns], axis=1)
    return data
```

```python
train_df = clean_data(train_df)
train_df
```

```python
test_df = clean_data(test_df)
test_df
```

# Split X,y from dataframe

```python
def Extract_from_data(data : pd.DataFrame):
    X = data.drop(columns = ['Survived'])
    y = data['Survived']
    return X, y
```

```python
X_train, y_train = Extract_from_data(train_df)
X_test , y_test  = Extract_from_data(test_df)
```

# Model 4 : TensorFlow

- **ModelCheckpoint** = ตัวช่วย save weights ที่ดีที่สุดระหว่างการทำงานได้
- **EarlyStopping** = ตัวช่วยตัดการเทรน model เมื่อ ค่าที่ต้องการเช่น loss ไม่ลดลงจากเดิมแล้วต่อเนื่อง patience epochs แล้วจะหยุดการทำงานทันทีจากนั้นจะโหลด weights เข้ามา ซึ่งเป็น weight ที่ดีที่สุดก่อนทำการ earlystop
- **ReduceLROnPlateau** = ตัวช่วยลดค่า learning rate เมื่อเทรนไปแล้ว patience epochs แต่ค่า loss ไม่ลดลงเลยจะปรับลด learning rate เพื่อเรียนรู้แบบละเอียดขึ้น
- ทั้ง 3 callbacks นี้สามารถแยกใช้กันได้หมด เลือกใช้อันใดอันหนึ่งก็ได้

```python
import tensorflow as tf

custom_callbacks = [
    tf.keras.callbacks.ModelCheckpoint(
        filepath="callback_save/callback_model",
        save_weights_only=True,
        monitor='val_accuracy',
        mode='max',
        save_best_only=True
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        mode = 'max',
        patience = 100,
        restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='accuracy',
        mode = 'max',
        factor=0.8,
        patience = 20,
    )
]
```

```python
tf.random.set_seed(888)

model_4 = tf.keras.Sequential([
        tf.keras.Input( shape=(X_train.shape[1],) ),  # input shape = n_features
        tf.keras.layers.Dense(100, activation = 'relu'),  # layer
        tf.keras.layers.Dense(1, activation = 'sigmoid'),   # output
], name = f"model_4")

model_4.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-3), # 1e-3
                loss = tf.keras.losses.binary_crossentropy,
                metrics = ['accuracy'],
)

history = model_4.fit(X_train_sc, y_train,
                    validation_data = (X_test_sc,y_test),
                    epochs = 500,
                    batch_size = 32,
                    callbacks = custom_callbacks,
            )
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.title(f"Model 4")
plt.grid()
plt.plot(history.history['accuracy'], label = "train_accuracy")
plt.plot(history.history['val_accuracy'], label = "val_accuracy")
plt.legend()
plt.show()
```

```python
loss, accuracy = model_4.evaluate(X_test_sc,y_test)
```

```python
print(f"Model_score : {round(accuracy*100,3)}/100")
```
